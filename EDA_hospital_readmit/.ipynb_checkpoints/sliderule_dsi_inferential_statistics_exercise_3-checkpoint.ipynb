# Hospital Readmissions Data Analysis and Recommendations for Reduction

### Background
In October 2012, the US government's Center for Medicare and Medicaid Services (CMS) began reducing Medicare payments for Inpatient Prospective Payment System hospitals with excess readmissions. Excess readmissions are measured by a ratio, by dividing a hospital’s number of “predicted” 30-day readmissions for heart attack, heart failure, and pneumonia by the number that would be “expected,” based on an average hospital with similar patients. A ratio greater than 1 indicates excess readmissions.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
#import and clean data
hospital_df = pd.read_csv('data/cms_hospital_readmissions.csv')

#remove non-data records
discharge_filter = hospital_df['Number of Discharges'] != 'Not Available'
xs_redmt_filter = hospital_df['Excess Readmission Ratio'].notnull()
hospital_df = hospital_df[discharge_filter & xs_redmt_filter]

#convert Discharges to int
hospital_df.loc[:, 'Number of Discharges'] = hospital_df['Number of Discharges'].astype(int)
```

```python
def pearson_r(x_data, y_data):
    """
    Takes x and y data arrays
    Returns the pearson r coefficient
    """
    corr_matrix = np.corrcoef(x_data, y_data)
    return corr_matrix[0,1]
```

```python
#get discharge and readmission series
discharges = hospital_df['Number of Discharges']
excess_readmit = hospital_df['Excess Readmission Ratio']

#calculate slope, Pearson r, and r-squared for discharges vs. readmission least squares line
slope, intercept = np.polyfit(discharges, excess_readmit, 1)
corr_coef = pearson_r(discharges, excess_readmit)
r_squared = corr_coef**2

print('Slope:', slope, '\t Intercept:', intercept)
print('Pearson R:', corr_coef)
print('R^2:', r_squared)
```

It is true that there is a negative relationship between hospital discharges and re-admission rate. But the correlation between the two is very small and a very small $R^{2}$ indicates that there is not a very strong fit to the data.

```python
#get dataframes for hospitals with low and high discharge rates
low_dis = hospital_df[hospital_df['Number of Discharges'] < 100]
high_dis = hospital_df[hospital_df['Number of Discharges'] > 1000]

#get mean readmission rate
low_rate = round(low_dis['Excess Readmission Ratio'].mean(),3)
high_rate = round(high_dis['Excess Readmission Ratio'].mean(),3)

#calculate % with readmission ratio > 1
low_excess = low_dis[low_dis['Excess Readmission Ratio']>1]
high_excess = high_dis[high_dis['Excess Readmission Ratio']>1]

low_pct = round(len(low_excess)/len(low_dis)*100)
high_pct = round(len(high_excess)/len(high_dis)*100)

print('The average excess readmission ratio for hospitals with less than 100 discharges is:',low_rate, 'and',low_pct,'% of hospitals have an excess readmission ration greater than 1')
print()
print('The average excess readmission ratio for hospitals with greater than 1000 discharges is:',high_rate, 'and',high_pct,'% of hospitals have an excess readmission ration greater than 1')
```

These numbers corroborate the statements above.

#### While it is true that there is a negative relationship between discharges and readmission rate, I do not believe that they have provided enough evidence to substantiate their claim. Why are they suggesting a discharge rate of 300 or fewer to monitor when their statistics never mention the number 300. Also, the fact that the correlation coefficient is so low indicates that the data may not actually support such measures.

## Hypothesis Test
### $H_{0}$: There is no relationship between number of discharges and excess readmissions 
### $H_{A}$: There is a relationship between number of discharges and excess readmissions

```python
#take bootstrap replicates and calculate R-squared
r_squared_reps = np.empty(10000)

for i in range(10000):
    dis_perm = np.random.permutation(discharges)
    r_squared_reps[i] = pearson_r(dis_perm, excess_readmit)**2

#calculate p-value
p = np.sum(r_squared_reps >= r_squared)/len(r_squared_reps)

print('p-value =', p)
```

A p-value of 0.0 indicates that none of the 10,000 bootstrap replicates had an $R_{2}$ as high as the one found in the original data. The means that at an alpha of 0.01 we would reject the null hypothesis that there is no relationship between discharge rate and excess readmission rate.

This is a case where statistical significance and practical significance don't line up. While, statisitically we can reject the statement that there is no relation ship between discharge and readmission, we have shown that the correlation is so weak that practically there might not be much of a difference. 

We can see this in the plot below. Which may be a better representation than the one provided in the original data. In this plot, you can see that the data are spread very wide, and most of the points are concentrated on the left side of the graph. This is much less misleading than the previously provided plot where we are fooled by the occurrence of points in the red and green boxes but are unable to understand how many points are concentrated in the white ares of the plot.

```python
#plot discharge vs readmission as well as least squares line
slope, intercept = np.polyfit(discharges, excess_readmit, 1)
ls_x = np.array([0,discharges.max()])
ls_y = slope*ls_x + intercept

plt.plot(discharges, excess_readmit, linestyle='none', marker = '.', alpha=0.2)
plt.plot(ls_x, ls_y)
plt.title('Number of Discharges vs. Excess Readmission Rate')
plt.xlabel('Number of Discharges')
plt.ylabel('Excess Readmission Rate')

plt.show()
print('Slope:', slope)
print('R-squared:', corr_coef**2)
```

```python
#get dataframes for each treatment type
df_knee = hospital_df[hospital_df['Measure Name'] == 'READM-30-HIP-KNEE-HRRP']
df_hf = hospital_df[hospital_df['Measure Name'] == 'READM-30-HF-HRRP']
df_copd = hospital_df[hospital_df['Measure Name'] == 'READM-30-COPD-HRRP']
df_pn = hospital_df[hospital_df['Measure Name'] == 'READM-30-PN-HRRP']
df_ami = hospital_df[hospital_df['Measure Name'] == 'READM-30-AMI-HRRP']

def get_least_squares(x, y):
    """
    Takes an x-data array and y-data array
    Returns x and y data for least squares line
    """
    slope, intercept = np.polyfit(x, y, 1)
    ls_x = np.array([0,x.max()])
    ls_y = slope*ls_x + intercept
    return ls_x, ls_y

#plot dicharges vs. readmission and least squares line for each treatment type
discharges_knee = df_knee['Number of Discharges']
excess_knee = df_knee['Excess Readmission Ratio']
knee_x, knee_y = get_least_squares(discharges_knee, excess_knee)
knee_r = pearson_r(discharges_knee, excess_knee)
knee_label = 'Knee: r= ' + str(round(knee_r,3))
plt.plot(discharges_knee, excess_knee, linestyle='none', marker='.',
            alpha=0.2, color='blue', label=knee_label)
plt.plot(knee_x, knee_y, color='blue')

discharges_hf = df_hf['Number of Discharges']
excess_hf = df_hf['Excess Readmission Ratio']
hf_x, hf_y = get_least_squares(discharges_hf, excess_hf)
hr_r = pearson_r(discharges_hf, excess_hf)
hf_label = 'HF: r= ' + str(round(hr_r,3))
plt.plot(discharges_hf, excess_hf, linestyle='none', marker='.',
            alpha=0.2, color='red', label=hf_label)
plt.plot(hf_x, hf_y, color='red')

discharges_copd = df_copd['Number of Discharges']
excess_copd = df_copd['Excess Readmission Ratio']
copd_x, copd_y = get_least_squares(discharges_copd, excess_copd)
copd_r = pearson_r(discharges_copd, excess_copd)
copd_label = 'COPD: r= ' + str(round(copd_r,3))
plt.plot(discharges_copd, excess_copd, linestyle='none', marker='.',
            alpha=0.2, color='green', label=copd_label)
plt.plot(copd_x, copd_y, color='green')

discharges_pn = df_pn['Number of Discharges']
excess_pn = df_pn['Excess Readmission Ratio']
pn_x, pn_y = get_least_squares(discharges_pn, excess_pn)
pn_r = pearson_r(discharges_pn, excess_pn)
pn_label = 'PN: r= ' + str(round(pn_r,3))
plt.plot(discharges_pn, excess_pn, linestyle='none', marker='.',
            alpha=0.2, color='black', label=pn_label)
plt.plot(pn_x, pn_y, color='black')

discharges_ami = df_ami['Number of Discharges']
excess_ami = df_ami['Excess Readmission Ratio']
ami_x, ami_y = get_least_squares(discharges_ami, excess_ami)
ami_r = pearson_r(discharges_ami, excess_ami)
ami_label = 'AMI: r= ' + str(round(pn_r,3))
plt.plot(discharges_ami, excess_ami, linestyle='none', marker='.',
            alpha=0.2, color='yellow', label=ami_label)
plt.plot(ami_x, ami_y, color='yellow')

plt.legend()
plt.title('Readmission Rate for Various Illnesses')
plt.xlabel('Number of Discharges')
plt.ylabel('Excess Readmession Rate')
plt.show()
```
